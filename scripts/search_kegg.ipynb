```python
import sys
sys.path.insert(0, "../src/")
from KEGGX import *
```

```python
from glob import glob

paths = glob('../data/human_KGML/*')
```

```python
dfs = []

for path in paths: 
    
    pathway_id = path.split('/')[-1].split('.')[0]
    pathway = KEGGX(path)
    
    compound_nodes_df = pathway.node_attributes_df[pathway.node_attributes_df['type'] == 'compound'].index
    
    if len(pathway.edge_attributes_df) > 0: 
        df = pathway.edge_attributes_df[(pathway.edge_attributes_df['source'].isin(indices)) | (pathway.edge_attributes_df['target'].isin(indices))]
        df['pathway'] = pathway_id
        dfs.append(df)
```

```python
all_edges = pd.concat(dfs)
```
